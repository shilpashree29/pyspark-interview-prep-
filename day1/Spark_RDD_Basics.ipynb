🔧 3. RDD Basics and Lambda Functions
✅ What is an RDD?
An RDD (Resilient Distributed Dataset) is the fundamental data structure in Spark, representing an immutable distributed collection of objects.

✅ Lambda Functions in PySpark
Lambda is an anonymous, one-line function often used in transformations like map, filter, and reduce.

🔁 4. RDD Transformations

4.1 map() – Apply function to each element

```
rdd = spark.sparkContext.parallelize([1, 2, 3, 4, 5])
rdd_map = rdd.map(lambda x: x * 2)
rdd_map.collect()  # [2, 4, 6, 8, 10]
```

4.2 flatMap() – Flatten the result

```
rdd = spark.sparkContext.parallelize(["hello world", "hi spark"])
rdd_flatmap = rdd.flatMap(lambda line: line.split(" "))
rdd_flatmap.collect()  # ['hello', 'world', 'hi', 'spark']
```

4.3 filter() – Filter elements

```
rdd = spark.sparkContext.parallelize([1, 2, 3, 4, 5])
rdd_filter = rdd.filter(lambda x: x % 2 == 0)
rdd_filter.collect()  # [2, 4]
```

4.4 distinct() – Remove duplicates

```
rdd = spark.sparkContext.parallelize([1, 2, 2, 3])
rdd_distinct = rdd.distinct()
rdd_distinct.collect()  # [1, 2, 3]
```

4.5 union() – Combine two RDDs

```
rdd1 = spark.sparkContext.parallelize([1, 2])
rdd2 = spark.sparkContext.parallelize([3, 4])
rdd_union = rdd1.union(rdd2)
rdd_union.collect()  # [1, 2, 3, 4]
```

4.6 intersection() – Common elements

```
rdd1 = spark.sparkContext.parallelize([1, 2, 3])
rdd2 = spark.sparkContext.parallelize([2, 3, 4])
rdd1.intersection(rdd2).collect()  # [2, 3]
```

🧮 5. Key-Value RDD Operations

5.1 reduceByKey() – Efficient aggregation

```
rdd = spark.sparkContext.parallelize([("a", 1), ("a", 2), ("b", 3)])
rdd.reduceByKey(lambda x, y: x + y).collect()  # [('a', 3), ('b', 3)]
```

5.2 groupByKey() – Group values (⚠️ less efficient)

```
rdd.groupByKey().mapValues(list).collect()  # [('a', [1, 2]), ('b', [3])]
```

✅ 6. RDD Actions

```
# RDD Actions

# Sample RDD
rdd = spark.sparkContext.parallelize([1, 2, 3, 4])

# 6.1 collect() – Brings all elements to the driver (avoid for large datasets)
print("collect():", rdd.collect())

# 6.2 count() – Number of elements
print("count():", rdd.count())

# 6.3 first() – First element
print("first():", rdd.first())

# 6.4 take(n) – First n elements
print("take(3):", rdd.take(3))

# 6.5 reduce() – Combine all elements using a function
print("reduce():", rdd.reduce(lambda x, y: x + y))

# 6.6 foreach() – Apply function to each element (prints on worker nodes)
# Note: Output might not appear in notebook UI depending on the Spark cluster behavior
rdd.foreach(lambda x: print(f"foreach(): {x}"))
```
